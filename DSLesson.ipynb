```python
import seaborn as sns
import matplotlib.pyplot as plt

# draw a countplot using seaborn
plt.style.available
```

```python
plt.style.use("seaborn")
```

```python
sns.countplot(x="Age",data=data,palette="magma") #Oranges,Blues,magma,flare,crest
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of members")
```

```python
# create a barplot
import numpy
sns.barplot(x="Balance",y="CreditScore",data=data,palette="magma",hue="Geography",estimator=numpy.mean)
plt.title("Distribution of Gender status")
plt.xlabel("Balance")
plt.ylabel("CreditScore")
# hue defines the categorical value
```
